```python
import numpy as np
```

```python
# Example data
data = [15, 20, 35, 40, 50]
```

```python
# Calculate the 90th percentile
percentile_90 = np.percentile(data, 90)
print(f'90th Percentile: {percentile_90}')
```

```python
# Calculate quartiles
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

print(f'Q1: {q1}, Q2 (Median): {q2}, Q3: {q3}')
```

```python
#Use box plots to visualize quartiles and percentiles.
import matplotlib.pyplot as plt
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
from scipy.stats import percentileofscore

# Example data
data = [15, 20, 35, 40, 50, 60, 65, 70, 80, 90]

# Target number
target = 80

# Calculate the percentile
percentile = percentileofscore(data, target, kind='rank')
print(f"The percentile of {target} is {percentile:.2f}%.")
```
